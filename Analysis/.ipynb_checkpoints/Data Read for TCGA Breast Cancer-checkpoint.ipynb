## TCGA Breast Cancer の　データの読み込み
　　データは、csvファイルとして、ローカルに保存されている。それを読み込む。

```python
import pandas as pd
import numpy as np
import pickle
import os
import xlrd
from scipy import stats
import csv
import numpy as np
import matplotlib.pyplot as plt


dat=pd.read_csv('C:/Users/maete/TCGA-Breast-Cancer-Biomarker/Raw Data/TCGA_BRCA.csv')
dat.head()


```

# Breast Cancer のサブタイプごとにデータを分割する。

```python
PAM50=dat.dropna(subset=['PAM50'])
TNBC=dat.dropna(subset=['TNBC'])
PAM50lite=dat.dropna(subset=['PAM50lite'])
```

## これらのデータをpickleデータとしてローカルに保存する。

```python
# Pickle として出力

with open('PAM50.pickle',mode='wb') as f:
    pickle.dump(PAM50,f)
    
with open('TNBC.pickle',mode='wb') as f:
    pickle.dump(TNBC,f)
    
with open('PAM50lite.pickle',mode='wb') as f:
    pickle.dump(PAM50lite,f)
        
```
